```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
def get_data(dir: str,
    num_tests: int = 19,
         acceleration: bool = True) -> tuple[list, list]:
    data = [0.0] * 16
    data[0] = 0.30672
    for i in range(2, num_tests):
        filename = f"results/{dir}/result{i+1}.log"
        with open(filename, "r") as file:
            for idx, line in enumerate(file):
                data[idx + 1] += float(line.split(" ")[1])
        
    num = [(i + 1) for i in range(len(data))]
    T_1 = data[0]
    data = [T_1] + [el / num_tests for el in data[1:]]
    print(data)
    
    if acceleration:
        T_1 = data[0]
        for idx in range(len(data)):
            data[idx] = T_1 / data[idx]
            
    return num, data
            
```

# №1

```python
num, data_reduction = get_data("1_1", acceleration=False)
_, data_no_reduction = get_data("1_2", acceleration=False)

fig, ax = plt.subplots()
ax.plot(num, data_reduction, label="reduction")
ax.plot(num, data_no_reduction, label="no reduction")
ax.legend(loc='upper left')
ax.set_xlabel('num threads')
ax.set_ylabel('time (s)')
ax.set_title('dependence of time(s) on the number of threads (N=1000000)')
```
